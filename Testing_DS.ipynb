"""Given a list of numbers, sort the list in ascending order"""

```python
def listsorter(old:list):
    """Sorts numbers in a list from min to max."""
    for j in range(len(old)):
        mini_num = old[j]
        count = j
        mini_idx = count

        #to get minimum/lowest number
        for i in range(j,len(old)):
            if old[i]<mini_num:
                mini_num = old[i]
                mini_idx = count
            count+=1

        #to swap
        tmp_hlod = old[j]
        old[j] = mini_num
        old[mini_idx] = tmp_hlod
    return old

call = [3,-4,0,5,80]
print(listsorter(call))
```

# Analysing Covid 19 Data

```python
import pandas as pd
data = pd.read_csv('covid_19_data.csv')
data.head() #to show the first 5 records by default
```

```python
data.drop(['SNo','Last Update'],axis=1,inplace=True) #to drop specific column or row-(axis=0)/ inplace=affect main data
data.head(10)
```

```python
data.rename(columns={'ObservationDate':'Date', 'Province/State':'Province', 'Country/Region':'Country'},inplace=True) #to rename data headings
data['Date'] = pd.to_datetime(data['Date']) #converting date format to internal pandas date format
data.head()
```

```python
# data.describe() #to show statistical description of data
# data.info() #info about the null 
data = data.fillna('NA') #to fill empty or null data 
```

```python
data.head(20)
```

```python
data2 = data.groupby(['Country','Date'])[['Country','Date','Confirmed','Deaths','Recovered']].sum().reset_index() #to group and summarize data
data2
```

```python
data3 = data2[data2['Recovered']>7] #to get data where recovered cases is > 7
data3
```

# Using MatPlotLib to visualize data

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
"""Testing matplotlib"""
x = np.linspace(0,10,1000) #start from 0 to 10 and generate 1000 points
y = np.sin(x)
plt.plot(x,y)#plot a simple line graph of x against y
```

```python
"""Plotting the 'confirmed, recovered, and deaths' data for each unique country"""
countries = data3['Country'].unique()
for i in countries:
    c = data3[data3['Country']==i].reset_index() #to get index of referred countries
    plt.scatter(np.arange(0,len(c)), c['Confirmed'], color='blue', label='Confirmed')
    plt.scatter(np.arange(0,len(c)), c['Recovered'], color='green', label='Recovered')
    plt.scatter(np.arange(0,len(c)), c['Deaths'], color='red', label='Deaths')
    plt.title(i)
    plt.xlabel('Days since the first suspect.')
    plt.ylabel('Number of cases')
    plt.legend()
    plt.show()
```

```python
"""Plotting the 'confirmed, recovered, and deaths' data for the world"""

data4 = data3.groupby(['Date'])[['Date','Confirmed','Deaths','Recovered']].sum().reset_index() #grouping all data by dates
c = data4
plt.scatter(np.arange(0,len(c)), c['Confirmed'], color='blue', label='Confirmed')
plt.scatter(np.arange(0,len(c)), c['Recovered'], color='green', label='Recovered')
plt.scatter(np.arange(0,len(c)), c['Deaths'], color='red', label='Deaths')
plt.title('World')
plt.xlabel('Days since the first suspect.')
plt.ylabel('Number of cases')
plt.legend()
plt.show()
```
